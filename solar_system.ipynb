# Let's make a diagram for solar system's planets

take planets' data from a CSV file `solar_system.csv`

use some helper-fuction from `solar_system_helpers.py` 

``` python 
    def get_data_from_csv(fn : str, df : pd.DataFrame) -> pd.DataFrame
    def gen_planets_as_circles(df: pd.DataFrame) -> (np.array, np.array)
    def gen_diagram(fig: plt.figure, ax: plt.axes, vals: np.array, df: pd.DataFrame) -> plt.figure
```

````python
import pandas as pd
import numpy as np
import datetime as dt

import matplotlib.pyplot as plt

import solar_system_helpers as my
````

## draw the diagram
planets are bigger than the distances from sun - dependent on the given exaggration factor

````python
def show_diagram(e_factor: int):
    fig, ax = plt.subplots(figsize=(20,3)) 
    # note we must use plt.subplots, not plt.subplot
    # (or if you have an existing figure)
    # fig = plt.gcf()
    # ax = fig.gca()

    # get the data from CSV
    df = my.get_data_from_csv('solar_system.csv', pd.DataFrame)
    # get planets as circels as an array for patches 
    # and some plantes' vals
    circles, vals = my.gen_planets_as_circles(df, e_factor)
    # generate the diagram
    my.gen_diagram(fig, ax, vals, df, e_factor)

    plt.suptitle("Solar system's planets", fontsize=12)
    plt.title(f'planets {e_factor:#3d} times bigger than the distances', fontsize=8)
    plt.show();
    return

show_diagram(e_factor=1000)
````

````python
import ipywidgets as w

w.interact(show_diagram,
           e_factor={'250': 250, '500': 500, '1.000': 1_000, '2.500': 2_500});
````
