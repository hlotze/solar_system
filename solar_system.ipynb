````python
fig, ax = plt.subplots(figsize=(20,3)) 
# note we must use plt.subplots, not plt.subplot
# (or if you have an existing figure)
# fig = plt.gcf()
# ax = fig.gca()

# get the data from CSV
df = my.get_data_from_csv('solar_system.csv', pd.DataFrame)
# get planets as circels as an array for patches 
# and some plantes' vals
circles, vals = my.gen_planets_as_circles(df)
# generate the diagram
my.gen_diagram(fig, ax, vals, df)

plt.suptitle("Solar system's planets", fontsize=12)
plt.title("planets 2.500 times bigger than the distances from sun", fontsize=8)
plt.show()
````
